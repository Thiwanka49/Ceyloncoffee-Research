```python
# imports
import pandas as pd
import numpy as np
import joblib
```

```python
# load models
labor_model = joblib.load("/content/drive/MyDrive/Coffee/Dev/Phase 1/C3/Labor/labor_demand_model.pkl")
labor_scaler = joblib.load("/content/drive/MyDrive/Coffee/Dev/Phase 1/C3/Labor/labor_feature_scaler.pkl")
```

```python
# User Daily Inputs
area_ha = 1.5
predicted_yield_kg_per_ha = 620
temp = 27.5
feelslike = 29.0
humidity = 78
precip = 12.0
severerisk = 10
month = 7
```

```python
# Internal feature engineering
daily_harvest_kg = (area_ha * predicted_yield_kg_per_ha) / 30

productivity_index = (
    1.0
    - (precip / 100) * 0.3
    - (humidity / 100) * 0.2
)
productivity_index = np.clip(productivity_index, 0.4, 1.0)

df_input = pd.DataFrame([{
    "area_ha": area_ha,
    "predicted_yield_kg_per_ha": predicted_yield_kg_per_ha,
    "daily_harvest_kg": daily_harvest_kg,
    "temp": temp,
    "feelslike": feelslike,
    "humidity": humidity,
    "precip": precip,
    "severerisk": severerisk,
    "productivity_index": productivity_index,
    "month": month
}])

FEATURE_ORDER = [
    "area_ha",
    "predicted_yield_kg_per_ha",
    "daily_harvest_kg",
    "temp",
    "feelslike",
    "humidity",
    "precip",
    "severerisk",
    "productivity_index",
    "month"
]

df_input = df_input[FEATURE_ORDER]
```

```python
# scale inputs
X_scaled = labor_scaler.transform(df_input)
```

```python
# inference
predicted_labor = labor_model.predict(X_scaled)[0]

pickers = int(round(predicted_labor[0]))
harvesters = int(round(predicted_labor[1]))
loaders = int(round(predicted_labor[2]))

```

```python
# output
print("Predicted Daily Labor Requirement")
print("--------------------------------")
print(f"Pickers     : {pickers}")
print(f"Harvesters  : {harvesters}")
print(f"Loaders     : {loaders}")
```
